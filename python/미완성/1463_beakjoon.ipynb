# 빠른 방법

```python
def f(n):
    if n<3: 
        return n-1
    else:
        # 1씩 빼는걸 나머지 만큼 시행 추가로 해석
        a = f(n//2)+n%2 
        b = f(n//3)+n%3
        return  1+min(a,b)
```

```python
f=lambda n:n-1if n<3 else 1+min(f(n//2)+n%2,f(n//3)+n%3)
print(f(int(input())))
```
